```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_pinguins = pd.read_csv("Anexo_pinguins-2.csv", index_col=False)
```

```python
df_pinguins
```

```python
num_linhas = df_pinguins.shape[0]

print('linhas: ', num_linhas)
```

```python
print(tabulate(df_pinguins.head(), headers='keys', tablefmt='pretty'))
```

```python
print("Colunas presentes:", df_pinguins.columns)
```

```python
# Verificar pinguins sem anotaÃ§Ãµes
pinguins_sem_anotacoes = df_pinguins[df_pinguins.isnull().any(axis=1)]

print(pinguins_sem_anotacoes)
```

```python
# Contar a frequÃªncia de ilhas
frequencia_ilhas = df_pinguins['Ilha'].value_counts()
print(frequencia_ilhas)
```

```python
# Contar a frequÃªncia das espÃ©cies
frequencia_especies = df_pinguins['Espece'].value_counts()
print(frequencia_especies)
```

```python
# Scatter plot para Largura do Bico vs. Profundidade do Bico colorido por EspÃ©cie
sns.scatterplot(data=df_pinguins, x='largura do bico', y='profundidade do bico', hue='Espece')
plt.show()

# Boxplot de Massa Corporal por EspÃ©cie
sns.boxplot(data=df_pinguins, x='Espece', y='massa corporal')
plt.show()
```

```python
# Boxplot de Largura do Bico por Sexo, dividido por EspÃ©cie
sns.boxplot(data=df_pinguins, x='Espece', y='largura do bico', hue='sexo')
plt.show()
```

```python
import matplotlib.pyplot as plt

# GrÃ¡fico de dispersÃ£o para largura do bico e profundidade do bico
plt.figure(figsize=(10, 6))
for especie in df_pinguins['Espece'].unique():
    subset = df_pinguins[df_pinguins['Espece'] == especie]
    plt.scatter(subset['largura do bico'], subset['profundidade do bico'], label=especie)

# ConfiguraÃ§Ãµes do grÃ¡fico
plt.xlabel('Largura do Bico (mm)')
plt.ylabel('Profundidade do Bico (mm)')
plt.title('RelaÃ§Ã£o entre Medidas do Pinguim e EspÃ©cie')
plt.legend()
plt.grid(True)
plt.show()
```
